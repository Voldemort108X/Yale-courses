```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Exercise 1
numOfSample = 500
x = np.random.normal(0, 0.1, numOfSample)
y = np.random.normal(0, 0.1, numOfSample)

def func_computeCovariance(x, y):
    numOfSample = x.shape[0]

    E_x = np.sum(x) / numOfSample
    E_y = np.sum(x) / numOfSample

    print(E_x.shape)

    return (x-E_x).T.dot(y-E_y)/ (numOfSample - 1)

def func_computeCovarianceMatrix2D(x, y):
    cov_xy = func_computeCovariance(x,y)
    cov_xx = func_computeCovariance(x,x)
    cov_yy = func_computeCovariance(y,y)
    return np.array([[cov_xx, cov_xy], [cov_xy, cov_yy]])
# x_axis = np.arange(0,10,1)

plt.figure()
plt.plot(x, y, '*')
plt.show()


print(func_computeCovarianceMatrix2D(x,y))
print(np.cov(x,y))

```
